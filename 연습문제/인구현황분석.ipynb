```python
import pandas as pd

data = pd.read_csv('../data/인구현황.csv')
data
```

```python
data.columns
```

```python
data1 = data.drop(data[data['행정기관'] == '전국'].index)
data1 = data1.loc[:, ['행정기관','총인구수']].sort_values(by='총인구수', ascending=False)
data1.head()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

#한글폰트 path 설정
font_path = 'C:\\windows\\Fonts\\malgun.ttf'
#font의 파일정보로 font name 을 알아내기
font_prop = fm.FontProperties(fname=font_path).get_name()
# matplotlib의 rc(run command) 함수를 사용해서 font name(Malgun Gothic) 설정
matplotlib.rc('font', family=font_prop)

figure,axes1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(25,12)
sns.barplot(data=data1, x='행정기관', y='총인구수', ax=axes1, hue='행정기관')
for item in axes1.get_xticklabels(): 
    item.set_rotation(45)
    item.set_fontsize(19)
for item in axes1.get_yticklabels():
    item.set_fontsize(19)
plt.rc('axes', labelsize=20)
plt.title(label='지역별 총인구수')
plt.rc('font', size=23)

plt.show()
```

```python
data2 = data.drop(data[data['행정기관'] == '전국'].index)
data2 = data2.loc[:, ['행정기관','세대당 인구']]\
.sort_values(by='세대당 인구', ascending=False)\
.reset_index(drop=True)
data2.head()
```

```python
figure,axes2 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(25,12)
sns.barplot(data=data2, x='행정기관', y='세대당 인구', ax=axes2, hue='행정기관')
for item in axes2.get_xticklabels(): 
    item.set_rotation(45)
    item.set_fontsize(19)
for item in axes2.get_yticklabels():
    item.set_fontsize(19)
plt.rc('axes', labelsize=20)
plt.title(label='지역별 세대당 인구')
plt.rc('font', size=23)

plt.show()
```

```python
data3 = data.drop(data[data['행정기관'] == '전국'].index)
data3 = data3.loc[:, ['행정기관','남여 비율']]\
.sort_values(by='남여 비율', ascending=False)\
.reset_index(drop=True)
data3
```

```python
figure,axes3 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(30,12)
sns.barplot(data=data3, x='행정기관', y='남여 비율', ax=axes3, hue='행정기관')
for item in axes3.get_xticklabels(): 
    item.set_rotation(45)
    item.set_fontsize(22)
for item in axes3.get_yticklabels():
    item.set_fontsize(22)
plt.rc('axes', labelsize=20)
plt.title(label='지역별 남녀 비율')
plt.rc('font', size=23)

# axhline을 legend(범례)로 지정하기
hline = plt.axhline(y=1.0, color='r', linestyle='--', linewidth=4)
hline.set_label('남녀 1:1 비율')
plt.legend(loc='upper right', fontsize='large', frameon=True)

plt.show()
```

```python
data_x = data.drop(data[data['행정기관'] == '전국'].index)
data_x = data_x['남여 비율']

for i in data_x.index:
    if data_x[i] == 1.0:
        data_x[i] = '동일'
    elif data_x[i] < 1.0:
        data_x[i] = '여초'
    else:
        data_x[i] = '남초'

data_x
```

```python
data_e = data.drop(data[data['행정기관'] == '전국'].index)
data4 = data_e.loc[:,['행정기관','남여 비율']].copy()
data4['남초여초'] = data_x
data4
```

```python
data5 = data2.loc[data2['세대당 인구'] > data2['세대당 인구'].mean()]
data5
```

```python
data_f = data.drop(data[data['행정기관'] == '전국'].index)\
.sort_values(by='세대수', ascending=False)
```

```python
figure, (axes1,axes2) = plt.subplots(nrows=2, ncols=1)
figure.set_size_inches(25,12)

sns.countplot(data=data4, x='남초여초', ax=axes1, order=data4['남초여초'].value_counts().index, hue='남초여초')
sns.barplot(data=data_f, x='행정기관', y='세대수', ax=axes2, hue='세대수')

axes1.set_title('남초 vs 여초 지역 개수')
axes2.set_title('지역별 세대수')

plt.rc('axes', labelsize=13)
plt.rc('figure', titlesize=13)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10) 

plt.show()
```
